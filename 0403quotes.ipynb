```python
from sklearn.feature_extraction.text import CountVectorizer

# Sample corpus
corpus = [
    'This is the first document.',
    'This document is the second document.',
    'And this is the third one.',
    'Is this the first document?'
]

# Initialize CountVectorizer
vectorizer = CountVectorizer()

# Fit and transform the corpus
X = vectorizer.fit_transform(corpus)

# Display the document-term matrix
print('Document-Term Matrix:')
print(X.toarray())

# Display the feature names (vocabulary)
print('\nFeature Names (Vocabulary):')
print(vectorizer.get_feature_names_out())
```

```python
import requests
from bs4 import BeautifulSoup

quotes_list = []
base_url = "http://quotes.toscrape.com/"

# Loop through pages 1 to 10
for page_num in range(1, 11):
    url = f"{base_url}page/{page_num}/"
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    # Extract all quotes from the current page
    quotes_on_page = soup.find_all('div', class_='quote')
    for quote_div in quotes_on_page:
        text = quote_div.find('span', class_='text').get_text()
        quotes_list.append(text)

print(f"Total quotes scraped: {len(quotes_list)}")
print("First 5 scraped quotes:")
for i, quote in enumerate(quotes_list[:5]):
    print(f"- {quote}")
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd

# Initialize CountVectorizer for the scraped quotes
# We might want to remove common English stop words to get more meaningful insights
vectorizer = CountVectorizer(stop_words='english')

# Fit and transform the quotes_list
X_quotes = vectorizer.fit_transform(quotes_list)

# Get feature names (words)
feature_names = vectorizer.get_feature_names_out()

# Sum the counts of each word across all documents
word_counts = X_quotes.sum(axis=0)

# Create a dictionary of word counts
counts_dict = dict(zip(feature_names, word_counts.tolist()[0]))

# Convert to a pandas Series and sort to find the most frequent words
word_frequencies = pd.Series(counts_dict)

# Get the top 5 most frequent words
top_5_words = word_frequencies.nlargest(5)

print('\nTop 5 most frequent words (excluding stop words):')
print(top_5_words)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd

# Initialize CountVectorizer with binary=True for presence/absence encoding
# We'll continue to remove English stop words for cleaner representation
vectorizer_binary = CountVectorizer(stop_words='english', binary=True)

# Fit and transform the quotes_list
X_quotes_binary = vectorizer_binary.fit_transform(quotes_list)

# Get feature names (words)
feature_names_binary = vectorizer_binary.get_feature_names_out()

# Create a DataFrame from the binary encoded matrix
df_binary_encoded = pd.DataFrame(X_quotes_binary.toarray(), columns=feature_names_binary)

print('\nBinary Encoded DataFrame (first 5 rows):')
display(df_binary_encoded.head())
```
